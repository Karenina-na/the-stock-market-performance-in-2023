```python
df = pd.read_csv('./data/000001_平安银行.csv')
df.head()
```

```python
# kp = df['开盘']
# sp = df['收盘']
# zg = df['最高']
# zd = df['最低']
# zdf = df['涨跌幅']
# zde = df['涨跌额']
# cj = df['成交量']
# cje = df['成交额']
# zf = df['振幅']
# hsl = df['换手率']
date = df['日期']

# 按季节取均值
df['季节'] = pd.to_datetime(date).dt.quarter
# 去除日期列
df = df.drop('日期', axis=1)
df_sea_avg = df.groupby('季节').mean()
df_sea_eve = df.groupby('季节').std()
```

```python
df_sea_avg.head(), df_sea_eve.head()
```

```python
def get_season_avg(df):
    df['季节'] = pd.to_datetime(df['日期']).dt.quarter
    df = df.drop('日期', axis=1)
    df_sea_avg = df.groupby('季节').mean()
    df_sea_eve = df.groupby('季节').std()
    return df_sea_avg, df_sea_eve
```

```python
all_file = os.listdir('./data')
all_file = [i.split('_') for i in all_file if i.endswith('.csv')]
```

```python
all_df_avg = pd.DataFrame()
all_df_eve = pd.DataFrame()
for code, name_csv in all_file:
    df = pd.read_csv(f'./data/{code}_{name_csv}')
    df_sea_avg, df_sea_eve = get_season_avg(df)
    df_sea_avg['code'] = code
    df_sea_avg['name'] = name_csv[:-4]
    df_sea_eve['code'] = code
    df_sea_eve['name'] = name_csv[:-4]
    # all_data_avg.append(df_sea_avg)
    # all_data_eve.append(df_sea_eve)
    all_df_avg = pd.concat([all_df_avg, df_sea_avg])
    all_df_eve = pd.concat([all_df_eve, df_sea_eve])
    print(f'{code}_{name_csv} done')
```

```python
# 合并eve和avg，eve的每列名字后面加上_std
all_df_eve.drop(['code', 'name'], axis=1, inplace=True)
all_df_eve.columns = [i + '_std' for i in all_df_eve.columns]
all_df = pd.concat([all_df_avg, all_df_eve], axis=1)
all_df.head()
```

```python
# 将code和name放到第一列、第二列
all_df = all_df[['code', 'name'] + [i for i in all_df.columns if i not in ['code', 'name']]]
all_df.head()
```

```python
# save
all_df.to_csv('./data/season_avg.csv', index=False)
```
